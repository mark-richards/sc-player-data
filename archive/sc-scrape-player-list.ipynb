```python
import pandas as pd
```

```python
def flatten_position(column, requested_field):
    pos_list = []
    for pos in column:
        pos_list.append(pos[requested_field])
    return ' '.join(pos_list)


df_sc = pd.read_json('inputs/SC_2020_player_data.json')
df_sc.columns = ["sc_id", "sc_first_name", "sc_last_name", "sc_team_id", "sc_price", "sc_previous_games", "sc_previous_average", "sc_previous_total", "sc_feed_id", "sc_active", "sc_positions", "sc_team"]
df_sc['sc_position'] = df_sc['sc_positions'].apply(lambda x: flatten_position(x, 'position'))
df_sc['sc_long_team'] = df_sc['sc_team'].apply(lambda x: x['name'])
df_sc['sc_team'] = df_sc['sc_team'].apply(lambda x: x['abbrev'])
df_sc.to_csv('inputs/2020 SC player list.csv')
df_sc
```
