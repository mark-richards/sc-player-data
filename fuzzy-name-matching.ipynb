```python
import pandas as pd
pd.set_option('display.max_columns', 999)
pd.set_option('display.max_rows', 30)

df_sc_players = pd.read_csv('inputs/2020 SC player list.csv')
df_sc_players['player_team_id'] = df_sc_players['SC Player Name'].str.upper() + '_' + df_sc_players['Team'].str.upper()
df_sc_players
```

```python
df_team_names = pd.read_csv('inputs/all_sources_team_names.csv')
df_fanfooty_players = pd.read_csv('inputs/2020_fanfooty_player_list_with_ids.csv')

df_fanfooty_players.columns = ['fanfooty_player_id', 'afl_player_id', 'first_name', 'surname', 'team', 'status', 'number', 'birth_date', 'height', 'weight', 'state_of_origin', 'recruited_from', 'games', 'goals']
df_fanfooty_players['short_team'] = df_fanfooty_players['team'].map(dict(df_team_names[['footywire_team_name', 'sc_team_name']].values))
df_fanfooty_players['player_team_id'] = df_fanfooty_players['first_name'].str.upper() + ' ' + df_fanfooty_players['surname'].str.upper() + '_' + df_fanfooty_players['short_team'].str.upper()
df_fanfooty_players
```

```python
merged = pd.merge(df_sc_players, df_fanfooty_players, how='left', left_on='SC Feed ID', right_on='afl_player_id')
merged.loc[merged['fanfooty_player_id'].isna()]
```
